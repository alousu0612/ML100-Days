### 用 iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('bmh')
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X = X[:, :-1]
features_names = iris.feature_names[:-1]
X.shape, y.shape, np.bincount(y)
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
def ploster(data, target, labels):
    fig = plt.figure(figsize=(8,6))
    ax = fig.gca(projection='3d')
    ax.scatter(data[:, 0], data[:, 1], data[:, 2],
               c=target, cmap=plt.cm.jet)
    ax.set_xlabel(labels[0])
    ax.set_ylabel(labels[1])
    ax.set_zlabel(labels[2])
    ax.view_init(elev=25, azim=45)
    

def km_est(n_clusters, data, n_init=10, init='k-means++'):
    est = KMeans(n_clusters=n_clusters)
    est.fit(data)
    return est.labels_
```

```python
# original data labels
ploster(X, y, features_names)
```

```python
# 2 clusters
ploster(X, km_est(2, X), features_names)
# 3 clusters
ploster(X, km_est(3, X), features_names)
# 4 clusters
ploster(X, km_est(4, X), features_names)
```

```python
# 2 clusters
ploster(X, km_est(2, X, n_init=1, init='random'), features_names)
# 3 clusters
ploster(X, km_est(3, X, n_init=1, init='random'), features_names)
# 4 clusters
ploster(X, km_est(4, X, n_init=1, init='random'), features_names)
```
