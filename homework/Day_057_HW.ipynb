### 用 iris (dataset.load_iris()) 資料嘗試跑 hierarchical clustering

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import AgglomerativeClustering
```

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
X = X[:, :-1]
features_names = iris.feature_names[:-1]
```

載入 相關套件 並 執行 hierarchical clustering 實驗 ...

```python
n_clusters = 3
```

```python
estimators = [('hierarchical_clustering_ward', 'ward'),
              ('hierarchical_clustering_complete', 'complete'),
              ('hierarchical_clustering_average', 'average')]
```

```python
def ploster(data, target, labels, title):
    fig = plt.figure(figsize=(8, 6))
    ax = Axes3D(fig, rect=[0, 0, 1, 1], elev=25, azim=40)
    
    ax.scatter(data[:, 0], data[:, 1], data[:, 2],
               c=target.astype(np.float), edgecolor='k')
    
    
    ax.set_xlabel(labels[0])
    ax.set_ylabel(labels[1])
    ax.set_zlabel(labels[2])
    ax.set_title(title)
    plt.show()
    
def ac_est(n_clusters, data, linkage):
    est = AgglomerativeClustering(n_clusters=n_clusters, linkage=linkage)
    est.fit(data)
    return est.labels_
```

```python
ploster(data=X, target=y, labels=features_names, title='ground_truth')

for est, linkage in estimators:
    target = ac_est(n_clusters=n_clusters, data=X, linkage=linkage)
    ploster(data=X, target=target, labels=features_names, title=est)
```
